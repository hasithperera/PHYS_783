```python
import scipy as sp
import numpy as np

import matplotlib.pyplot as plt



dat_0 = sp.io.loadmat('./exp_8_0_data_f.mat')
dat_1 = sp.io.loadmat('./exp_8_1_data_f.mat')
```

```python
vx_ax = dat_0['v_ax'][0,:]
x_ax = dat_0['x_ax'][0,:]
t_ax = dat_0['t_ax'][0,:]
v_phi = dat_0['v_phi'][0,0]

f1_0 = dat_0['f1']

## get exp 2 data

f1_1 = dat_1['f1']
f3_1 = dat_1['f3']
```

```python
print(f1_0.shape)
print(t_ax.shape)
print(v_phi)
```

```python

y_lim = .03

plt.plot(vx_ax,np.sum(f1_1[:,:,-1],axis=1)-np.sum(f1_1[:,:,0],axis=1),'g',label='$e_0$ with bump')
plt.plot(vx_ax,np.sum(f3_1[:,:,-1],axis=1)-np.sum(f3_1[:,:,0],axis=1),'m',label='$e_1$ with bump')
plt.plot(vx_ax,np.sum(f1_0[:,:,-1],axis=1)-np.sum(f1_0[:,:,0],axis=1),'--k',label='landau damping')


plt.plot([v_phi,v_phi],[-y_lim,y_lim],':r',label='$v_\phi$')
plt.legend(loc='upper right')
plt.xlim([0,10])
plt.xlabel('$v/v_{th}$')
plt.ylim([-.02,0.02])
plt.title('$\int{f-f_0} dx$')
```
