```python
import scipy as sp
import numpy as np

import matplotlib.pyplot as plt



dat_0 = sp.io.loadmat('./exp_4_0_data.mat')
dat_2 = sp.io.loadmat('./exp_4_2_data.mat')
```

```python
t_0 = dat_0['t'][0,:]
t_2 = dat_2['t'][0,:]

plt.figure(figsize=(7,3))

plt.plot(t_0,dat_0['fld'][0,:],'g',label='$E_{fld}$')
plt.plot(t_2,dat_2['fld'][0,:],'--g',label="_")

plt.plot(t_0,dat_0['part'][0,:],'k',label='$E_{particles}$')
plt.plot(t_2,dat_2['part'][0,:],'--k')

plt.plot(t_0,dat_0['total'][0,:],'m',label='$E_{t}$')
plt.plot(t_2,dat_2['total'][0,:],'.-m',markevery=2.5)

plt.xlabel('t')
plt.legend(loc='upper left',bbox_to_anchor=(0,-.2),ncols=3)

f = plt.gcf()
f.dpi = 300

ax = plt.gca()
ax.ticklabel_format(scilimits=(-1,2))
```
